```python
#1
hdfc_ohlc=hdfc[['Open','High','Low','Close']].copy()
hdfc_ohlc
```

```python
hdfc_ohlc['date']=pd.to_datetime(hdfc_ohlc.index)
hdfc['date']=pd.to_datetime(hdfc.index)
```

```python
#2
hdfc_ohlc.date=pd.to_datetime(hdfc_ohlc['date'])
hdfc_mean=hdfc_ohlc.groupby(hdfc_ohlc.date.dt.year)[['Open','High','Close','Low']].mean()
hdfc_mean

```

```python
#3
hdfc_ohlc['date']=pd.to_datetime(hdfc_ohlc['date']) - pd.to_timedelta(7, unit='d')
hdfc_mean1=hdfc_ohlc.groupby([pd.Grouper(key='date', freq='W-MON'),hdfc_ohlc.date.dt.year])[['Open','High','Close','Low']].mean()
hdfc_mean1
```

```python
#4
infy_volume=hdfc.groupby([hdfc.date.dt.year,hdfc.date.dt.month])[['Volume']].sum()
infy_volume.head(15)
```

```python
infy_percen=hdfc.groupby([hdfc.date.dt.year,hdfc.date.dt.month])[['Volume']].apply(lambda x: x.diff().div(x).mul(100))
infy_percen
```

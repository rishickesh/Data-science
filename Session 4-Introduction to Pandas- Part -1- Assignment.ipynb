```python
# 7. Here is a dictionary of short names for all the team names in the dataset. Create a new dataframe called 
# matches_alias by replacing names in all the columns with long team names.
team_names = {'Kolkata Knight Riders': 'KKR', 
              'Chennai Super Kings': 'CSK', 
              'Rajasthan Royals' : 'RR',
               'Mumbai Indians' : 'MI', 
              'Deccan Chargers' : 'DC', 
              'Kings XI Punjab' : 'KXIP',
               'Royal Challengers Bangalore' : 'RCB', 
              'Delhi Daredevils' : 'DD',
              'Kochi Tuskers Kerala' : 'KTK', 
              'Pune Warriors' : 'PW', 
              'Sunrisers Hyderabad' : 'SRH',
              'Rising Pune Supergiant' : 'RPS', 
              'Gujarat Lions' : 'GL'}
matches_alias=matches
matches_alias["winner"].replace(team_names, inplace=True)
matches_alias["team1"].replace(team_names, inplace=True)
matches_alias["team2"].replace(team_names, inplace=True)
matches_alias["toss_winner"].replace(team_names, inplace=True)
matches_alias
```

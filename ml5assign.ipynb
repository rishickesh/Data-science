```python
# import the 'Breast Cancer' dataset
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
data = pd.read_csv('C://Users/User/Downloads/breastc.csv')
data.head()
```

```python
# Create 2 clusters for the dataset using KMeans clustering
from sklearn.cluster import KMeans
from sklearn import datasets 
data = datasets.load_breast_cancer()
X = data.data
y = data.target
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
kmeans.labels_
kmeans.cluster_centers_  
```

```python
# visualise the KMeans clusters
colors = ['red', 'blue']
pred_colors = [colors[label] for label in kmeans.labels_]
actual_colors = [colors[label] for label in y]
plt.scatter(X[:,0], X[:, 29], c=pred_colors)
plt.title('KMeans clusters')
```

```python
plt.scatter(X[:,0], X[:, 29], c=actual_colors)
plt.title('KMeans clusters')
```

```python
# Create Hierarchicial clustering model using linkages : {“ward”, “complete”, “average”}
from sklearn.cluster import AgglomerativeClustering
n_clusters = 2
ward = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
ward.fit(X)
colors = c=np.random.rand(n_clusters, 3)
pred_colors = [colors[label] for label in ward.labels_]
plt.scatter(X[:, 0], X[:, 1], c=pred_colors)
plt.title('Hierarchical clustering')
```

```python
from sklearn.cluster import AgglomerativeClustering
n_clusters = 2
ward = AgglomerativeClustering(n_clusters=n_clusters, linkage='complete')
ward.fit(X)
colors = c=np.random.rand(n_clusters, 3)
pred_colors = [colors[label] for label in ward.labels_]
plt.scatter(X[:, 0], X[:, 1], c=pred_colors)
plt.title('Hierarchical clustering')
```

```python
# visualise the clusters
from sklearn.cluster import AgglomerativeClustering
n_clusters = 2
ward = AgglomerativeClustering(n_clusters=n_clusters, linkage='average')
ward.fit(X)
colors = c=np.random.rand(n_clusters, 3)
pred_colors = [colors[label] for label in ward.labels_]
plt.scatter(X[:, 0], X[:, 1], c=pred_colors)
plt.title('Hierarchical clustering')
```

```python
# Compare the results
 #within cluster sum of squares
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) 
    
plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('wcss')
```
